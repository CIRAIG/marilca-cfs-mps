Processing of the regional marine currents data to get regional averages 

```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv('Velocities_all.csv')

# Density mapping
density_map = {
    'EPS': 1,
    'HDPE': 4,
    'LDPE': 3,
    'PET': 6,
    'PP': 2,
    'PS': 5
}

# Add density code column
df['Density_code'] = df['Polymer_all'].map(density_map)

# Define regions and their countries with weights
regions = {
    'Northern regions': {
        'KaraSea': 7.96E+02
    },
    'Central Asia': {
        'SeaOkhotsk': 3.83E+04
    },
    'Southeast Asia': {
        'EC-Japan': 1.47E+06,
        'BayBengal': 7.28E+04
    },
    'Oceania': {
        'Oceania': 1.53E+03
    },
    'Europe': {
        'Greece': 7.32E+01,
        'NorthSea': 4.73E+02
    },
    'Africa & Middle East': {
        'Nile': 8.48E+04,
        'GG_Ext': 3.52E+04,
        'Congo': 1.34E+04
    },
    'North America': {
        'NorthAmerica': 5.72E+02
    },
    'Latin America': {
        'SouthAmerica': 3.73E+03
    }
}

# Columns that need weighted averaging (only v_ and k_ columns)
weighted_columns = [
    'v_Surf2Col_cont_median_all',
    'v_Col2Sed_cont_median_all',
    'k_Surf2glo_median_all',
    'k_Col2glo_median_all',
    'v_Surf2Col_glo_median_all',
    'v_Col2Sed_glo_median_all'
]

# Create a list to store results
results = []

# Group by Polymer, Density, Shape, Size, Diameter
grouping_cols = ['Polymer_all', 'Density_all', 'Shape_all', 'Size_all', 'Diameter_all']

for group_keys, group_df in df.groupby(grouping_cols):
    # Process each region
    for region_name, country_weights in regions.items():
        row_data = {
            'Region': region_name,
            'Polymer_all': group_keys[0] if isinstance(group_keys, tuple) else group_keys,
            'Density_all': group_keys[1] if isinstance(group_keys, tuple) else None,
            'Shape_all': group_keys[2] if isinstance(group_keys, tuple) else None,
            'Size_all': group_keys[3] if isinstance(group_keys, tuple) else None,
            'Diameter_all': group_keys[4] if isinstance(group_keys, tuple) else None,
            'Density_code': group_df.iloc[0]['Density_code']
        }
        
        # Track which countries had data for this combination
        countries_used = []
        
        # Calculate weighted averages for each column
        for col in weighted_columns:
            total_weight = 0
            weighted_sum = 0
            
            for country, weight in country_weights.items():
                # Find matching country in this group
                country_data = group_df[group_df['Country_all'] == country]
                
                if not country_data.empty:
                    value = country_data[col].iloc[0]
                    # Only include if value is not NaN
                    if np.isfinite(value):
                        weighted_sum += value * weight
                        total_weight += weight
                        if country not in countries_used:
                            countries_used.append(country)
            
            # Calculate weighted average (handle case where all values were NaN)
            if total_weight > 0:
                row_data[col] = weighted_sum / total_weight
            else:
                row_data[col] = 0
        
        # Add the list of countries used
        row_data['Country_all'] = ';'.join(countries_used) if countries_used else ''
        
        results.append(row_data)

# ----------------------------------------------------------------------
# ADD NEW REGION: "Default USEtox" – simple (unweighted) average per group
# ----------------------------------------------------------------------
for group_keys, group_df in df.groupby(grouping_cols):
    row_data = {
        'Region': 'Default USEtox',
        'Polymer_all': group_keys[0] if isinstance(group_keys, tuple) else group_keys,
        'Density_all': group_keys[1] if isinstance(group_keys, tuple) else None,
        'Shape_all': group_keys[2] if isinstance(group_keys, tuple) else None,
        'Size_all': group_keys[3] if isinstance(group_keys, tuple) else None,
        'Diameter_all': group_keys[4] if isinstance(group_keys, tuple) else None,
        'Density_code': group_df.iloc[0]['Density_code'],
        'Country_all': ''   # No country list for the default region
    }
    
    # Calculate simple mean for each weighted column (ignoring NaN)
    for col in weighted_columns:
        # Replace inf and -inf with NaN, then drop NaNs
        finite_vals = group_df[col].replace([np.inf, -np.inf], np.nan).dropna()
        if len(finite_vals) > 0:
            row_data[col] = finite_vals.mean()
        else:
            row_data[col] = 0
    
    results.append(row_data)

# Create result dataframe
result_df = pd.DataFrame(results)

# Define the exact column order as requested
column_order = [
    'Region',
    'Country_all',
    'Polymer_all',
    'Density_code',
    'Density_all',
    'Shape_all',
    'Size_all',
    'Diameter_all',
    'v_Surf2Col_cont_median_all',
    'v_Col2Sed_cont_median_all',
    'k_Surf2glo_median_all',
    'k_Col2glo_median_all',
    'v_Surf2Col_glo_median_all',
    'v_Col2Sed_glo_median_all'
]

# Ensure all columns exist (fill missing with 0)
for col in column_order:
    if col not in result_df.columns:
        result_df[col] = 0

result_df = result_df[column_order]

# Replace any remaining NaN with 0
result_df = result_df.fillna(0)

# Custom sorting order for regions (original + Default USEtox at the end)
region_order = [
    'Northern regions',
    'Central Asia',
    'Southeast Asia',
    'Oceania',
    'Europe',
    'Africa & Middle east',
    'North America',
    'Latin America',
    'Default USEtox'          # New region appears last
]

# Convert Region to categorical with custom order
result_df['Region'] = pd.Categorical(result_df['Region'], categories=region_order, ordered=True)

# Sort the dataframe
result_df = result_df.sort_values(
    by=['Region', 'Polymer_all', 'Shape_all', 'Size_all'],
    ascending=[True, True, True, True]
)

# Reset index after sorting
result_df = result_df.reset_index(drop=True)

# Save to Excel
result_df.to_excel('Velocities_all_weighted.xlsx', index=False, sheet_name='Weighted_Velocities')

print(f"Processing complete!")
print(f"Created {len(results)} rows (combinations of polymer properties × regions)")
print(f"Output saved to: Velocities_all_weighted.xlsx")

# Display first few rows as preview
print("\nPreview of first 10 rows:")
print(result_df.head(10).to_string())

# Show unique regions to verify ordering
print("\nRegions in output (in order):")
print(result_df['Region'].unique())
```

```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook

# ------------------------------
# 1. Read the weighted Excel file
# ------------------------------
file_path = 'Velocities_all_weighted.xlsx'
df = pd.read_excel(file_path, sheet_name='Weighted_Velocities')

# Conversion factor: days to seconds
day_to_sec = 24 * 60 * 60

# ------------------------------
# 2. Prepare output list for long format
# ------------------------------
long_rows = []

# Base columns for grouping
base_cols = ['Region', 'Polymer_all', 'Density_code', 'Diameter_all', 'Shape_all']

# ---- A) cont columns (v_ only, Regional scale) ----
cont_columns = {
    'v_Surf2Col_cont_median_all': ('Regional', 'sea', 'VelInput'),
    'v_Col2Sed_cont_median_all': ('Regional', 'deepocean', 'VelInput')
}

for _, row in df.iterrows():
    region = row['Region']
    for col, (scale, subcomp, varname) in cont_columns.items():
        value = row[col]
        if np.isfinite(value):
            long_rows.append({
                'region': region,
                'polymer': row['Polymer_all'],
                'Density code': row['Density_code'],
                'size': row['Diameter_all']*1000000,
                'shape': row['Shape_all'],
                'Scale': scale,
                'SubCompart': subcomp,
                'varName': varname,
                'Waarde': value / day_to_sec
            })

# ---- B) k_ glo columns (Regional scale, no global average) ----
k_columns = {
    'k_Surf2glo_median_all': ('Regional', 'sea', 'AdvInput'),
    'k_Col2glo_median_all': ('Regional', 'deepocean', 'AdvInput')
}

for _, row in df.iterrows():
    region = row['Region']
    for col, (scale, subcomp, varname) in k_columns.items():
        value = row[col]
        if np.isfinite(value):
            long_rows.append({
                'region': region,
                'polymer': row['Polymer_all'],
                'Density code': row['Density_code'],
                'size': row['Diameter_all']*1000000,
                'shape': row['Shape_all'],
                'Scale': scale,
                'SubCompart': subcomp,
                'varName': varname,
                'Waarde': value / day_to_sec
            })

# ---- C) v_ glo columns (Continental scale, global simple average) ----
# Compute global average per polymer group
v_glo_columns = {
    'v_Surf2Col_glo_median_all': ('Continental', 'sea', 'VelInput'),
    'v_Col2Sed_glo_median_all': ('Continental', 'deepocean', 'VelInput')
}

group_cols = ['Polymer_all', 'Density_code', 'Diameter_all', 'Shape_all']
grouped = df.groupby(group_cols)

for (polymer, density_code, size, shape), group in grouped:
    for col, (scale, subcomp, varname) in v_glo_columns.items():
        # Extract finite values (ignore NaN/inf)
        finite_vals = group[col].replace([np.inf, -np.inf], np.nan).dropna()
        if len(finite_vals) > 0:
            mean_val = finite_vals.mean()
        else:
            mean_val = 0.0
        long_rows.append({
            'region': 'Global',
            'polymer': polymer,
            'Density code': density_code,
            'size': size * 1000000,
            'shape': shape,
            'Scale': scale,
            'SubCompart': subcomp,
            'varName': varname,
            'Waarde': mean_val / day_to_sec
        })

# ------------------------------
# 3. Create long format DataFrame
# ------------------------------
long_df = pd.DataFrame(long_rows)

# Sort for readability
long_df = long_df.sort_values(
    by=['region', 'polymer', 'shape', 'size', 'Scale', 'SubCompart']
).reset_index(drop=True)

# ------------------------------
# 4. Write to a new sheet in the same Excel file
# ------------------------------
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    long_df.to_excel(writer, sheet_name='LongFormat', index=False)

print(f"Long format sheet added to {file_path}")
print(f"Total rows in long format: {len(long_df)}")
print("\nPreview of the first 10 rows:")
print(long_df.head(10).to_string())
```

```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook

# ------------------------------
# 1. Read the weighted Excel file
# ------------------------------
file_path = 'Velocities_all_weighted.xlsx'
df = pd.read_excel(file_path, sheet_name='Weighted_Velocities')

# Conversion factor: days to seconds
day_to_sec = 24 * 60 * 60

# ------------------------------
# 2. Prepare output list for long format
# ------------------------------
long_rows = []

# Base columns for grouping
base_cols = ['Region', 'Polymer_all', 'Density_code', 'Diameter_all', 'Shape_all']

# ---- A) cont columns (v_ only, Regional scale) ----
cont_columns = {
    'v_Surf2Col_cont_median_all': ('Regional', 'sea', 'VelInput'),
    'v_Col2Sed_cont_median_all': ('Regional', 'deepocean', 'VelInput')
}

for _, row in df.iterrows():
    region = row['Region']
    for col, (scale, subcomp, varname) in cont_columns.items():
        value = row[col]
        if np.isfinite(value):
            long_rows.append({
                'region': region,
                'polymer': row['Polymer_all'],
                'Density code': row['Density_code'],
                'size': row['Diameter_all']*1000000,
                'shape': row['Shape_all'],
                'Scale': scale,
                'SubCompart': subcomp,
                'varName': varname,
                'Waarde': value / day_to_sec
            })

# ---- B) k_ glo columns (Regional scale, no global average) ----
k_columns = {
    'k_Surf2glo_median_all': ('Regional', 'sea', 'AdvInput'),
    'k_Col2glo_median_all': ('Regional', 'deepocean', 'AdvInput')
}

for _, row in df.iterrows():
    region = row['Region']
    for col, (scale, subcomp, varname) in k_columns.items():
        value = row[col]
        if np.isfinite(value):
            long_rows.append({
                'region': region,
                'polymer': row['Polymer_all'],
                'Density code': row['Density_code'],
                'size': row['Diameter_all']*1000000,
                'shape': row['Shape_all'],
                'Scale': scale,
                'SubCompart': subcomp,
                'varName': varname,
                'Waarde': value / day_to_sec
            })

# ---- C) v_ glo columns (Continental scale, global simple average) ----
# Compute global average per polymer group
v_glo_columns = {
    'v_Surf2Col_glo_median_all': ('Continental', 'sea', 'VelInput'),
    'v_Col2Sed_glo_median_all': ('Continental', 'deepocean', 'VelInput')
}

group_cols = ['Polymer_all', 'Density_code', 'Diameter_all', 'Shape_all']
grouped = df.groupby(group_cols)

for (polymer, density_code, size, shape), group in grouped:
    for col, (scale, subcomp, varname) in v_glo_columns.items():
        # Extract finite values (ignore NaN/inf)
        finite_vals = group[col].replace([np.inf, -np.inf], np.nan).dropna()
        if len(finite_vals) > 0:
            mean_val = finite_vals.mean()
        else:
            mean_val = 0.0
        long_rows.append({
            'region': 'Global',
            'polymer': polymer,
            'Density code': density_code,
            'size': size * 1000000,
            'shape': shape,
            'Scale': scale,
            'SubCompart': subcomp,
            'varName': varname,
            'Waarde': mean_val / day_to_sec
        })

# ------------------------------
# 3. Create initial long format DataFrame
# ------------------------------
long_df = pd.DataFrame(long_rows)

# ------------------------------
# 4. Add missing polymers by density code matching
# ------------------------------
# Define reference polymers and their density codes (from original data)
reference_polymers = {
    'EPS': 1,
    'HDPE': 4,
    'LDPE': 3,
    'PET': 6,
    'PP': 2,
    'PS': 5
}

# Define all polymers and their density codes (including new ones)
all_polymers_mapping = {
    'EPS': 1,
    'HDPE': 4,
    'LDPE': 3,
    'PET': 6,
    'PP': 2,
    'PS': 5,
    'PA_Nylon': 6,
    'PVC': 6,
    'TRWP': 6,
    'PHA': 6,
    'PLA': 6,
    'PBSA': 6,
    'starch_blend': 6,
    'PBAT': 6,
    'Cotton': 6,
    'Linen': 6,
    'Viscose': 6,
    'Rayon': 6,
    'Lyocell': 6,
    'Modal': 6,
    'PU': 5,
    'PAN': 5,
    'PCL': 5,
}

# Get existing polymers in long_df
existing_polymers = set(long_df['polymer'].unique())

# Prepare list for new rows
new_rows = []

# For each density code, find the reference polymer (must exist in existing_polymers)
for density_code in set(reference_polymers.values()):
    # Find reference polymer for this density code
    ref_polymer = None
    for poly, code in reference_polymers.items():
        if code == density_code:
            ref_polymer = poly
            break
    if ref_polymer is None or ref_polymer not in existing_polymers:
        continue
    
    # Get all target polymers with same density code that are NOT already existing
    target_polymers = [poly for poly, code in all_polymers_mapping.items() 
                       if code == density_code and poly not in existing_polymers]
    
    if not target_polymers:
        continue
    
    # Extract all rows for the reference polymer
    ref_rows = long_df[long_df['polymer'] == ref_polymer].copy()
    
    # For each target polymer, create copies of ref_rows with polymer name changed
    for target in target_polymers:
        new = ref_rows.copy()
        new['polymer'] = target
        # Also update Density code (though same, but ensure)
        new['Density code'] = density_code
        new_rows.append(new)

# Concatenate new rows if any
if new_rows:
    long_df = pd.concat([long_df] + new_rows, ignore_index=True)

# ------------------------------
# 5. Sort for readability
# ------------------------------
long_df = long_df.sort_values(
    by=['region', 'polymer', 'shape', 'size', 'Scale', 'SubCompart']
).reset_index(drop=True)

# ------------------------------
# 6. Write to a new sheet in the same Excel file
# ------------------------------
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    long_df.to_excel(writer, sheet_name='LongFormat', index=False)

print(f"Long format sheet added to {file_path}")
print(f"Total rows in long format: {len(long_df)}")
print(f"Polymers in output: {sorted(long_df['polymer'].unique())}")
print("\nPreview of the first 10 rows:")
print(long_df.head(10).to_string())
```
